```python
import pandas as pd
import certifi
import ssl
import geopy.geocoders
import time

from geopy.geocoders import OpenCage, ArcGIS, Photon, PickPoint
from geopy.extra.rate_limiter import RateLimiter
```

```python
# I dont understand the fix I got from stackoverflow but this fixes the SSL certification issue
ctx = ssl.create_default_context(cafile=certifi.where())
geopy.geocoders.options.default_ssl_context = ctx

file = './csv/ON Physicians.csv'
df = pd.read_csv(file)
```

```python
# Refer to geopy documentation for more info on geoencoders I used

# The encoder I used that was very good, but try ArcGIS if you have an account
photonCoder = Photon()
```

```python
# Gets the GPS coordinates from the search query given
def getGPSLocation(searchTerm):
    try:
        # location will equal an object returned from photon
        # Search term has Canada at the end to fence our query in this country
        # Timeout is 2.5s to wait for the server to respond before we say it failed connection
        location = photonCoder.geocode(query=(searchTerm+', Canada'), timeout=2.5)
        return location
    except:
        print('Failed')
        # return none if unable to do the search
        return None
```

```python
# Loop through all the rows in the list
for i in range(len(df)):
    
    # Get the location from the photon API
    location = getGPSLocation(str(df.loc[i, 'Street'])+', '+ str(df.loc[i, 'City']))
    
    # We limit our request to 1 per second due to api restrictions
    time.sleep(1)
    
    # if a none is returned then retry again this time with the postal code
    if location is None:
        location = getGPSLocation(str(df.loc[i, 'Postal Code'])+ df.loc[i, 'City'])
        time.sleep(1)
    
    # if location is none again then try the street + Province instead
    if location is None:
        location = getGPSLocation(str(df.loc[i, 'Street'])+', Ontario')
        time.sleep(1)
        
    # if location is none again then try Postal Code + Province instead (Final Attempt)
    if location is None:
        location = getGPSLocation(str(df.loc[i, 'Postal Code'])+', Ontario')
        time.sleep(1)
        
    # if we did get the location object then add it to the row under  the new cols
    try:
        print('{}, {}'.format(location.latitude, location.longitude))
        df.loc[i, 'Latitude'] = location.latitude
        df.loc[i, 'Longitude'] = location.longitude
    except:
        # If no location after all the attempts then leave it blank
        df.loc[i, 'Latitude'] = ''
        df.loc[i, 'Longitude'] = ''
```

```python
df.to_csv(file)
```
