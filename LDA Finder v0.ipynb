```python
import pandas as pd
import time
import geopandas as gpd

from geopandas.geoseries import GeoSeries
from shapely.geometry import Point, Polygon
```

```python
file = './csv/ON Pharmacy.csv'
df = pd.read_csv(file)
```

```python
'''
Download the LDA data from the stats canada boundary section. 
Download it in ArcGis sharp format to be compatible with geopands.
The coordinates in that file are in Lambert conic Coordinate system 
and needs to be converted to Mercantor GPS coordinate system. At the 
time of writing this, geopandas builtin converter was broken in Anaconda distribution. 
So use either pip to install geopanda or QGIS to convert the LDA to mercantor projection
and then load it in this notebook.
'''
# Load the boundaries polygons file in geopandas NOT regular pandas
disseminationAreas = gpd.read_file('./LDA/LDA GPS.shp')

# Broken code 
# disseminationAreas.crs = {'init' :'epsg:3347'}
# disseminationAreas = disseminationAreas.to_crs({'init': 'epsg:4326'})

# Dictionary of province ID that contains the LDAs
pruids = {
    'NF': '10',
    'PEI': '11',
    'NS': '12',
    'NB': '13',
    'QUE': '24',
    'ON': '35',
    'MA': '46',
    'SK': '47',
    'AL': '48',
    'BC': '59',
}
```

```python
# gets the lda points by comparing the GPS of the pharmacy/physician's practice as a point
# and sees if it's in any of the LDA polygons
def getLDAforPoints(df, pruid):
    
    # Get the index of the LDAs that are within the desired province
    index = disseminationAreas.index[disseminationAreas['PRUID'] == pruid].tolist()
    
    # Loop through each row of the pharmacy/physician list
    for i in range(len(df)):
        
        # Loop through the LDAs in the desired province that might contain the pharmacy/physician
        for j in index:
            
            # Convert the GPS coordinates into a geopy point first
            point = Point(float(df.loc[i, 'Latitude']), float(df.loc[i, 'Longitude']))
            
            # Check if that point is within the LDA
            if point.within(disseminationAreas.loc[j,'geometry']):
                print('we have a match')
                # if it is then set that rows LDA IDto that ID
                df.loc[i, 'DAUID'] = disseminationAreas.loc[j, 'DAUID']
            
    return df
```

```python
df = getLDAforPoints(df, pruids['ON'])
```

```python
df.to_csv(file, index=False)
```
