```python
import pandas as pd
import re
```

```python
file = './csv/ON Physicians.csv'
df = pd.read_csv(file)

# List of cities in ON
cities = pd.read_csv('./res/ON Cities.csv')
```

```python
# Uses REGEXP to extract the postal code
def extractPostalCode(x):
    pattern = r"[a-zA-z]\d[a-zA-z]\s\d[a-zA-z]\d"
    try: 
        result = re.search(pattern, x).group()
    except:
        result = "None"
    return  result


# searches the string to see if there is a match the the cities list
def extractCity(x):
    arrayString = x.split('\n')
    for string in arrayString:
        if 'ON' in string:
            return(string.split('ON')[0].rstrip())
        
    return ""

#
def extractStreet(x):
    arrayString = x.split('\n')
    for string in arrayString:
        if 'ON' in string:
            street = ' '.join(arrayString[:arrayString.index(string)])
            return(street.replace('\r', '').replace('\n', ''))
        
    return ""
```

```python
# Remove any non practicisng members
df = df[df['Member Type'] == 'Practising']

# Reset the index b/c we removed a lot of rows
df.reset_index(inplace=True)

# Extract the postal code from the Address 
df['Postal Code'] = df['Address'].map(extractPostalCode)

# Extract the City the Address 
df['City'] = df['Address'].map(extractCity)

# Drop the previous index and the middle name
df.drop(columns={'index', 'Middle Name', 'Address', 'Member Type', 'Register'}, inplace=True)
```

```python
df.to_csv(file, index=False)
```

```python
df.drop_duplicates(subset='Name', keep='first')
df.drop_duplicates(keep='first')
```

```python
# Extract the postal code from the Address 
df['Postal Code'] = df['Location'].map(extractPostalCode)

# Extract the City the Address 
df['City'] = df['Location'].map(extractCity)

# Extract the City the Address 
df['Street'] = df['Location'].map(extractStreet)
```

```python
df['First Name'] = df['Name'].map(lambda x : x.split(',')[1].split('(')[0])
df['Last Name'] = df['Name'].map(lambda x : x.split(',')[0])
```
